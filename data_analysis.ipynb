```python
data=pd.read_excel('/content/Applications_for_Machine_Learning_internship_edited.xlsx')
```

In this cell
1. add index to empty cells of column Name
2.deleting rows with 0 value in any of four python, machine, NLP, Deep Learning
3.find the average and convert to percentage

```python
data['Name']=range(1,len(data)+1)
bcolumns=['Python (out of 3)','Machine Learning (out of 3)','Natural Language Processing (NLP) (out of 3)','Deep Learning (out of 3)']
pcolumns=['Performance_UG','Performance_12','Performance_10']
data=data.dropna(subset=pcolumns)
data = data[~(data[bcolumns] == 0).any(axis=1)]
data['Average']=data[['Python (out of 3)','Machine Learning (out of 3)','Natural Language Processing (NLP) (out of 3)','Deep Learning (out of 3)']].mean(axis=1)
data['Average'] = (data['Average'] / 3) * 100
```

In this cell
1. Deleting values rows with null values in performance_UG,Performance_12 Performance_10
2.Normalizing values values that are out of 10 converting them to out of 100
3. Removing values after '/' because there are many errors in them
4. Then finding average of performances in three feilds (performance pg is not that much necessary and does not contain that much weightage)
5. Finding average of performance and skills
6. Putting them in decending order

  **TOP MOST PERSON IS IDEAL FOR INTERNSHIP**

```python
def transform_values(column):
    transformed_column = column.copy()
    for i, value in column.iteritems():
        if '/' in value:
            parts = value.split('/')
            transformed_column[i] = parts[0]
        elif float(value) < 10:
            transformed_column[i] = str(float(value) * 10)
    return transformed_column

columns_to_transform = ['Performance_UG', 'Performance_12', 'Performance_10']
data[columns_to_transform] = data[columns_to_transform].apply(transform_values)


# Convert values less than 10 to float and multiply by 10 for 'Performance_UG'
data.loc[data['Performance_UG'].astype(float) < 10, 'Performance_UG'] = (data['Performance_UG'].astype(float) * 10).astype(str)

# Convert values less than 10 to float and multiply by 10 for 'Performance_12'
data.loc[data['Performance_12'].astype(float) < 10, 'Performance_12'] = (data['Performance_12'].astype(float) * 10).astype(str)

# Convert values less than 10 to float and multiply by 10 for 'Performance_10'
data.loc[data['Performance_10'].astype(float) < 10, 'Performance_10'] = (data['Performance_10'].astype(float) * 10).astype(str)

# Compute the average of the three columns and store it in a new column 'Average'
data['perf_Average'] = data[['Performance_UG', 'Performance_12', 'Performance_10']].astype(float).mean(axis=1)


data['order'] = (data['perf_Average'] + data['Average']) / 2
data_sorted = data.sort_values(by='order', ascending=False)

data_sorted
```
